```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#http://archive.ics.uci.edu/ml/datasets/Concrete+Compressive+Strength
df = pd.read_excel('Concrete_Data.xls')
```

```
df
```

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df)
input_data_scaled = scaler.transform(df)
```

```
input_data_scaled
```

```
from minisom import MiniSom
som = MiniSom(x= 13, y= 13, input_len= 9, sigma= 1.0, learning_rate= 0.5, random_seed=0)
```

```
som.random_weights_init(input_data_scaled)
```

```
som.train_random(data = input_data_scaled, num_iteration= 100)
```

```
graph = som.distance_map()
```

```
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(graph, annot=True, linewidths=0, ax=ax, square=True, cmap='jet')
```
